```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df18_23 = pd.read_csv('../Data/ProducedData/refined_df18_23.csv')
```

### Alcohol Use and Poverty Level ###

```python
df18_23.groupby('POVERTY3')['ALDAYPWK'].mean()
```

```python
df18_23.groupby(['YEAR','POVERTY3'])['ALDAYPWK'].mean()
```

```python
grouped = df18_23.groupby(['YEAR','POVERTY3'])['ALDAYPWK'].mean().reset_index()

pivot_df = grouped.pivot(index='YEAR', columns='POVERTY3', values='ALDAYPWK')

styles = {
    1.0: {"color": "green",  "marker": "o", "linestyle": "-"},
    2.0: {"color": "red",   "marker": "s", "linestyle": "--"},
    3.0: {"color": "blue", "marker": "D", "linestyle": ":"},
}

plt.figure(figsize=(12,6))

for col in pivot_df.columns:
    style = styles.get(col, {})
    plt.plot(
        pivot_df.index,
        pivot_df[col],
        label=f"Poverty {col}",
        linewidth=2,
        markersize=7,
        **style
    )

plt.title("Mean Days Drank Alcohol Per Week by Poverty Level Over Time", fontsize=16)
plt.xlabel("Year", fontsize=14)
plt.ylabel("Mean ALDAYPWK", fontsize=14)
plt.legend(title="POVERTY3")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

plt.savefig('../Images/mean_days_drank_alc_per_wk_by_poverty_over_time.png')
```
