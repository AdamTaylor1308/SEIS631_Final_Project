```python
import pandas as pd
```

Reading in files and taking care of dtype errors:

```python
file_p18 = '/Users/adamtaylor/Documents/GradSchool/DataPreparationAnalysis/Final Project/GitHub Repo/SEIS631_Final_Project/Data/RawData/NSDUH_2018_Tab.tsv'
df18 = pd.read_csv(file_p18, sep='\t', dtype={2525: str})
file_p19 = '/Users/adamtaylor/Documents/GradSchool/DataPreparationAnalysis/Final Project/GitHub Repo/SEIS631_Final_Project/Data/RawData/NSDUH_2019_Tab.txt'
df19 = pd.read_csv(file_p19, sep='\t', dtype={2575: str})
file_p20 = '/Users/adamtaylor/Documents/GradSchool/DataPreparationAnalysis/Final Project/GitHub Repo/SEIS631_Final_Project/Data/RawData/NSDUH_2020_Tab.txt'
df20 = pd.read_csv(file_p20, sep='\t', dtype={2726: str})
file_p21_23 = '/Users/adamtaylor/Documents/GradSchool/DataPreparationAnalysis/Final Project/GitHub Repo/SEIS631_Final_Project/Data/RawData/NSDUH_2021_2023_Tab.txt'
df21_23 = pd.read_csv(file_p21_23, sep='\t', dtype={2283: str})
```

Filtering dataframes for desired columns:

```python
gdf18 = df18.loc[:,['IRPINC3','IRFAMIN3','IRWRKSTAT','POVERTY3','ALCYRTOT','ALDAYPWK','AMHTXRC3','MHRCOST2','SPDMON','SMIPP_U']] #'RCVMHASPTX'
gdf18['YEAR'] = 2018
```

```python
gdf19 = df19.loc[:, ['IRPINC3','IRFAMIN3','IRWRKSTAT','POVERTY3','ALCYRTOT','ALDAYPWK','AMHTXRC3','MHRCOST2','SPDMON','SMIPP_U']] #'RCVMHASPTX'
gdf19['YEAR'] = 2019
```

```python
gdf20 = df20.loc[:,['IRPINC3','IRFAMIN3','IRWRKSTAT','POVERTY3','ALCYRTOT','ALDAYPWK','AMHTXRC3','MHRCOST2','SPDMON','SMIPP_U']] 
gdf20['YEAR'] = 2020
```

```python
gdf21_23 = df21_23.loc[:,['YEAR','IRPINC3','IRFAMIN3','IRWRKSTAT','POVERTY3','ALCYRTOT','ALDAYPWK','IRMHTPHVID','MHTUNCOST','SPDPSTMON','SMIPPPY']]
gdf21 = gdf21_23[gdf21_23['YEAR'] == 2021]
gdf22 = gdf21_23[gdf21_23['YEAR'] == 2022]
gdf23 = gdf21_23[gdf21_23['YEAR'] == 2023]
```
