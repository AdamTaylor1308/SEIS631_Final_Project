```python
import pandas as pd
import numpy as np
```

```python
df18_23 = pd.read_csv('../Data/ProducedData/cleaned_gdf18_23.csv')
```

Values in `ALCYRTOT` should be in the range 0-365. 

```python
bad_list = []
for v in df18_23['ALCYRTOT'].unique():
    if v < 0:
        bad_list.append(v)
    elif v > 365:
        bad_list.append(v)
    
bad_list
```

Values in `ALDAYPWK`, `IRPINC3`, `IRFAMIN3` should be in the range 0-7. 

```python
cols = ['ALDAYPWK', 'IRPINC3', 'IRFAMIN3']
for col in cols:
    bad_list = []
    for v in df18_23[col].unique():
        if v < 0 or v > 7:
            bad_list.append(v)
    print(f"Column: {col}")
    print(f"Invalid values: {bad_list}\n")
```

```python
cols = ['ALCYRTOT']
for col in cols:
    bad_list = []
    for v in df18_23[col].unique():
        if v < 0 or v > 365:
            bad_list.append(v)
    print(f"Column: {col}")
    print(f"Invalid values: {bad_list}\n")
```

Filtering out rows of respondents who do not drink alcohol (`ALCYRTOT` values > 365, `ALDAYPWK` values > 7): 

```python
df18_23 = df18_23[df18_23['ALCYRTOT'].isin(range(0,365))]
df18_23 = df18_23[df18_23['ALDAYPWK'].isin(range(0,7))]
```

Saving the refined df18_23:

```python
refined_gdf18_23 = df18_23
refined_gdf18_23.to_csv('../Data/ProducedData/refined_df18_23.csv')
```
